```python
#Problem3

import pymysql.cursors
import re
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='root',
                             db='examen2',
                             charset='utf8mb4',  # note utf8... this isimportant for unusual characters!
                             cursorclass=pymysql.cursors.DictCursor)

def writedb(db, Locusgene, gene, ProteinLength):
    Locusgene = str(Locusgene)
    gene = str(gene)
    ProteinLength = str(ProteinLength)
    try:
        with db.cursor() as cursor:
            sql = "insert into Locusgene (Locusgene, gene, ProteinLength) values ('{}','{}', '{}')".format(Locusgene, gene, ProteinLength)
            cursor.execute(sql)
            db.commit()
    except Exception as e: 
        print("FALLO")
        return False
    finally: 
        return True

Locusgenef = open("LocusGene.tsv", "r")
Locusgenef.seek(0)
for line in Locusgenef.readlines():
    Locusgene, gene, ProteinLength = line.split ('\t')
    if writedb(connection, Locusgene, gene, ProteinLength):
        print("written {} {} {}".format(Locusgene, gene, ProteinLength))
    else:
        print("write failed")

Locusgenef.close()

def writedb2(db, Locus, germplasm, phenotype, pubmed):
    Locus = str(Locus)
    germplasm = str(germplasm)
    phenotype = str(phenotype)
    pubmed = str(pubmed)
    try:
        with db.cursor() as cursor:
            sql = "insert into germplasm (Locus, germplasm, phenotype, pubmed) values ('{}','{}', '{}', '{}')".format(Locus, germplasm, phenotype, pubmed)
            cursor.execute(sql)
            db.commit()
    except Exception as e: 
        print("FALLO")
        return False
    finally: 
        return True

Germplasm = open("Germplasm.tsv", "r")
Germplasm.seek(0)
for line in Germplasm.readlines():
    Locus, germplasm, phenotype, pubmed = line.split ('\t')
    if writedb2(connection, Locus, germplasm, phenotype, pubmed):
        print("written {} {} {} {}".format(Locus, germplasm, phenotype, pubmed))
    else:
        print("write failed")

Germplasm.close()
connection.close()

```
